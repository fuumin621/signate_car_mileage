```python
def split_name_2(text,num):
    split=text.split()
    try:
        return split[num]
    except:
        return 'No Label'
```

```python
df['car name 0']=df['car name'].apply(lambda x:split_name_2(text=x,num=0))
```

```python
df['car name 1']=df['car name'].apply(lambda x:split_name_2(text=x,num=1))
df['car name 2']=df['car name'].apply(lambda x:split_name_2(text=x,num=2))
```

```python
df['car name 0+1']=df['car name 0']+df['car name 1']
```

```python
df['car name 0+1+2']=df['car name 0']+df['car name 1']+df['car name 2']
```
