```python
import numpy as np
import pandas as pd
```

```python
df_car_name=pd.read_csv('df_car_name.csv')
```

```python
df_car_name.head()
```

```python
df_poly=pd.read_csv('df_poly.csv')
```

```python
df_poly.head()
```

```python
poly_only_columns=[i for i in df_poly.columns if i not in df_car_name.columns]
```

```python
poly_only_columns.append('id')
```

```python
df_poly_only=df_poly[poly_only_columns]
```

```python
df_poly_only.head()
```

```python
df=pd.merge(df_car_name,df_poly_only,on='id')
```

```python
df.to_csv('df_car_poly_merge.csv',index=None)
```
